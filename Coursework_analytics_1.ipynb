```python
# Ипортируем нужные библиотеки и считываем данные.
import pandas as pd
import numpy as np
course_contents = pd.read_csv("Course_Data\course_contents.csv")
courses = pd.read_csv("Course_Data\courses.csv")
progress_phases = pd.read_csv("Course_Data\progress_phases.csv")
progresses = pd.read_csv("Course_Data\progresses.csv")
students = pd.read_csv("Course_Data\students.csv")
```

```python
# переименуем колонки, чтобы не запутаться в одинаковых названиях
courses.columns='course_'+courses.columns
progresses.columns = ['progress_id', 'student_id', 'course_id']
students.columns='student_'+students.columns
```

```python
# соединим таблицы
main = pd.merge(how='outer',left=courses, right=progresses, left_on='course_id', right_on='course_id')
# main = pd.merge(how='outer',left=main, right=students, left_on='student_id', right_on='student_id')
    main = pd.merge(how='outer',left=main, right=course_contents, left_on='course_id', right_on='course_id')
main = pd.merge(how='outer',left=main, right=progress_phases, 
                left_on=['progress_id','module_number','lesson_number'], 
                right_on=['progress_id','module_number','lesson_number'])

# main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'].to_csv('compare.csv')
# не понятно, как сверять, проверочный сет не полный. По нескольким строкам вроде совпадает.
```

```python
# количество учеников на каждом курсе
students_count = pd.DataFrame(main[(main['status']=='start')
#            &(main['course_title']=='Excel Базовый')
          ]
       .groupby(['course_title','student_id']).count()['course_id']
       .groupby(['course_title']).count())

print('Количество учеников на каждом курсе:',
       students_count)
```

```python
# минимальный, максимальный, средний, медианный возраст студентов
from dateutil.relativedelta import relativedelta
birthday_dates = students.dropna(subset=['student_birthday'])
#'1988-07-16'
#Сконвертируем с заменой кривых дат на NaT
to_datetime = lambda x: pd.to_datetime(x, format='%Y-%m-%d',errors='coerce')
birthday_dates['student_birthday'] = birthday_dates['student_birthday'].apply(to_datetime)
# Теперь снова очистим данные
birthday_dates = birthday_dates.dropna(subset=['student_birthday'])
#Посчитаем возраст
f = lambda x: relativedelta(pd.to_datetime('10.03.2020', format='%d.%m.%Y'),x).years
birthday_dates['age']=birthday_dates['student_birthday'].apply(f)
#Уберем слишком большие и слишком маленькие возрасты
birthday_dates = birthday_dates[(birthday_dates['age']>5)]
birthday_dates = birthday_dates[(birthday_dates['age']<100)]
print(birthday_dates.head(50))
```

```python
print('Минимальный, максимальный, средний, медианный возраст студентов:',
      (birthday_dates['age'].min()),
      (birthday_dates['age'].max()),
      (birthday_dates['age'].mean()),
      (birthday_dates['age'].median())
       )

# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
with_ages = pd.merge(how='left',left=birthday_dates, right=main, left_on='student_id', right_on='student_id')
print('Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе:',
                                    with_ages
                                    .groupby(['course_title']).agg(['min','max','mean','median'])['age']
      )
```
